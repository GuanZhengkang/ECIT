```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable

# 参数设置
alpha = 0.9  # Levy稳定分布的稳定指数 (0 < alpha <= 2)
beta = 0  # 偏态参数 (-1 <= beta <= 1)
loc = 0  # 位置参数
scale = 1  # 缩放参数
n = 10  # 缩放因子

# 定义 x 范围
x = np.linspace(-5, 5, 1000)

# 计算原始 CDF 和缩放后 CDF
cdf_original = levy_stable.cdf(x, alpha, beta, loc=loc, scale=scale)
cdf_scaled = levy_stable.cdf(x, alpha, beta, loc=loc, scale=scale * (n ** (1 / alpha - 1)))

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, cdf_original, label="Original CDF", color="blue")
plt.plot(x, cdf_scaled, label=f"Scaled CDF (n={n})", color="red", linestyle="--")
plt.title("CDF of Levy Stable Distribution", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("CDF", fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
