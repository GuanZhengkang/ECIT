```python
from ensemble.simu import simu, simus
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 10
ensembles=[None,"cauchy"]
x = [200,400,800,1600,2400,3200,4000,5600]
dfs = []
for n in x:
    k = int(n/200)
    re = simus(n,t,k=k,ensembles=ensembles)
    df = re.copy()
    new_row = df.sum(axis=0)
    df.loc[len(df)] = new_row
    df.index = df.index[:-1].tolist() + ['SUM'] 
    dfs.append(df)
```

```python
col1_values = [df.iloc[0, 0] for df in dfs]
col2_values = [df.iloc[0, 1] for df in dfs]

plt.figure(figsize=(20, 6),dpi=100)
plt.plot(x, col1_values, label="None", marker="o")  # col1
plt.plot(x, col2_values, label="cauchy", marker="s")  # col2
plt.title("TypeI")
plt.legend()
plt.grid(True)
plt.show()
```
