```python
sns.set()

datasets = [kcitDZ1, kcitDZ2, kcitDZ3, kcitDZ4, kcitDZ5]
dz = list(range(1,6))

        
alpha = 0.95
markersize = 3.8
linewidth = 1.45


fig, axes = plt.subplots(1, 2, figsize=(7, 3), sharey=True, dpi=500)



###
kIs = []
e2s = []
e175s = []
rIs = []
kPs = []
e2Ps = []
e175Ps = []
rPs = []

# n = 1200
ni = 1
for data in datasets:
    kI = data[0][0][ni]
    e2I = data[1][0][ni]
    e175I = data[2][0][ni]
    rI = data[3][0][ni]

    kP = 1-data[0][1][ni]
    e2P = 1-data[1][1][ni]
    e175P = 1-data[2][1][ni]
    rP = 1-data[3][1][ni]

    kIs.append(kI)
    e2s.append(e2I)
    e175s.append(e175I)
    rIs.append(rI)

    kPs.append(kP)
    e2Ps.append(e2P)
    e175Ps.append(e175P)
    rPs.append(rP)


linestyles = ['--', ':', '-', '-.']  # 
markers = ['^', 'o', 's', 'D']      # 
colors = ["#cf444d", "#ff6969", sns.color_palette("muted")[0], sns.color_palette("muted")[9]]  # 



axes[0].plot(dz, kIs, alpha=alpha, label='KCIT', linestyle=linestyles[0], marker=markers[0], markersize=markersize, linewidth=linewidth, color=colors[0])
axes[0].plot(dz, rIs, alpha=alpha, label='RCIT', linestyle=linestyles[1], marker=markers[1], markersize=markersize, linewidth=linewidth, color=colors[1])
axes[0].plot(dz, e175s, alpha=alpha, label=r"EKCIT ($\alpha=1.75$)", linestyle=linestyles[2], marker=markers[2], markersize=markersize, linewidth=linewidth, color=colors[2])
axes[0].plot(dz, e2s, alpha=alpha, label=r"EKCIT ($\alpha=2$)", linestyle=linestyles[3], marker=markers[3], markersize=markersize, linewidth=linewidth, color=colors[3])


axes[1].plot(dz, kPs, alpha=alpha, label='KCIT', linestyle=linestyles[0], marker=markers[0], markersize=markersize, linewidth=linewidth, color=colors[0])
axes[1].plot(dz, rPs, alpha=alpha, label='RCIT', linestyle=linestyles[1], marker=markers[1], markersize=markersize, linewidth=linewidth, color=colors[1])
axes[1].plot(dz, e175Ps, alpha=alpha, label=r"EKCIT ($\alpha=1.75$)", linestyle=linestyles[2], marker=markers[2], markersize=markersize, linewidth=linewidth, color=colors[2])
axes[1].plot(dz, e2Ps, alpha=alpha, label=r"EKCIT ($\alpha=2$)", linestyle=linestyles[3], marker=markers[3], markersize=markersize, linewidth=linewidth, color=colors[3])



ax_eI, ax_eII = axes

ax_eI.set_xticks(dz)
ax_eII.set_xticks(dz)
ax_eI.set_ylim(0, 1)
ax_eII.set_ylim(0, 1)

ax_eI.axhline(y=0.05, color='black', linestyle='-', alpha=0.6, linewidth = 0.5)

ax_eI.set_title("Type I Error", fontsize=12)
ax_eI.set_ylabel("Error Rate", fontsize=11)

ax_eII.set_title("Power", fontsize=12)
ax_eII.set_ylabel("Power", fontsize=11)


ax_eII.legend(loc='lower right', fontsize=8.5, ncol=1)


ax_eI.set_xlabel("Dimension of Z", fontsize=11)
ax_eII.set_xlabel("Dimension of Z", fontsize=11)


plt.tight_layout()
plt.show()
```
